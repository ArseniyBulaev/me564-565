```python
import numpy as np
import scipy as cp
import matplotlib.pyplot as plt
```

```python
dx = 0.005
L = 1
x = np.linspace(dx, L, int((L - dx) / dx))
f = np.ones(len(x))
f[:int(len(f)/2)] = 0
plt.plot(x, f)
```

```python
x
```

```python
fFS = np.zeros(len(x))
A0 = (2/L) * np.sum(f)*dx
fFS = A0/2
N = 100
for n in range(1, N):
    An = (2/L)*np.sum(f*np.cos(2*np.pi*n*x/L))*dx
    Bn = (2/L)*np.sum(f*np.sin(2*np.pi*n*x/L))*dx
    fFS = fFS + An*np.cos(2*np.pi*n*x/L) + Bn*np.sin(2*np.pi*n*x/L)
```

```python
plt.plot(x, f)
plt.plot(x, fFS, 'r--')
plt.show()
```
