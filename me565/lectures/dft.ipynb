```python
import numpy as np
import scipy as cp
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
N = 500
x = np.ones(N)
w = np.exp(2j*np.pi/N)

y = cp.fft.fft(x)
```

```python
# slow and steady
# DFT = np.zeros((N,N),dtype=complex)
# for i in range(N):
#     for j in range(N):
#         DFT[i,j] = w**((i - 1) * (j - 1))

# fast
n = np.linspace(1,N,N)
I,J  = np.meshgrid(n,n)
DFT = w**((I - 1) * (J - 1))
```

```python
plt.matshow(DFT.real)
```

```python
plt.plot(np.real(y), 'k', label="built in")
plt.plot(np.real(DFT@x), 'r--' , label="our")
plt.legend()
```
