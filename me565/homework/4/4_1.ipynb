```python
import numpy as np
import scipy as cp
import matplotlib.pyplot as plt
```

```python
def f(x):
    if x < - 1 or x > 1 : return 0
    else : return 1 - np.abs(x) 
```

```python
x = np.linspace(-2,2,100)
y = [f(x_i) for x_i in x]
plt.plot(x, y)
```

```python
def a(n):
    return (4.0/(np.pi*n)**2) * (1 - np.cos(np.pi*n/2))

def b(n):
    return 0

def fFS(x, N):
    FS_sum = 1.0/4
    for n in range(1, N+1):
        FS_sum += a(n)*np.cos(np.pi*n*x/2) + b(n)*np.sin(np.pi*n*x/2) 
    return FS_sum
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(10,10)
ax.plot(x, y, "r--", label="f(x)")
for n in range(1, 11, 1):
    ax.plot(x, fFS(x, n), label="FS of f(x) with {} modes".format(n))
ax.legend()
```

```python
# Mode coefficients an and bn
N = 10
fig,ax = plt.subplots(N//5, 5,)
fig.set_figheight(10)
fig.set_figwidth(15)

for n in range(0,N,1):
    ax[n // 5, n%5].plot(x, a(n+1)**np.cos(np.pi*(n+1)*x/2))
    ax[n // 5, n%5].title.set_text("a_{}".format(n+1))
```

```python
N = 100
fig, ax = plt.subplots()
n = np.arange(1,N)
ax.plot(n, a(n), color="orange")
plt.title("First {} Fourier coefs".format(N))
```
