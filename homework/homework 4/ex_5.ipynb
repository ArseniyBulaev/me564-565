```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
epsilons = [0.01,0.1, 1, 20]
t0 = 0
tbound = 30
first_step = 0.1
y0 = [0.1, -1]
y_fix = [0, 0]
t = np.arange(t0, tbound, first_step)

def SYS(t, Y):
    return np.array([Y[1], -eps*Y[1]*(Y[0]**2 - 1) - Y[0]])
```

```python
fig, axs = plt.subplots(1, len(epsilons))
fig.set_figheight(5)
fig.set_figwidth(15)

for i,eps in enumerate(epsilons):
    sol = sp.integrate.solve_ivp(SYS, (t0,tbound), y0, t_eval=t).y.T
    axs[i].plot(y_fix[0], y_fix[1], "ro")
    axs[i].plot(sol[:, 0], sol[:, 1])
    axs[i].set_xlabel("y")
    axs[i].set_title("epsilon={}".format(eps))
axs[0].set_ylabel("y'")
plt.show()
```
