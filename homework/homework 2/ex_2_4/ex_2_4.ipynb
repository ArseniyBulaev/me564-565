```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
# Matrix
A = [
    [0, 1, 0],
    [0, 0, 1],
    [2, 1, -2]
]
```

```python
eigen_values, eigen_vectors = np.linalg.eig(A)
eigen_values
```

```python
# Boundary conditions consts system
C = [
    [1, 1, 1],
    [-2, -1, 1],
    [4, 1, 1]
]

y1 = np.array([1, 1, 1]).T
y2 = np.array([1,-1, 1]).T
```

```python
np.linalg.solve(C, y1)
```

```python
np.linalg.solve(C, y2)
```

```python
t0 = 0
tbound = 10
first_step = 0.01
max_step = first_step
t = np.arange(t0, tbound, first_step)
def  dxdt(t,X):
              return A @ X
```

```python
linear_ode_solution_0 = sp.integrate.solve_ivp(dxdt, (t0, tbound), y1, t_eval=t)
linear_ode_solution_1 = sp.integrate.solve_ivp(dxdt, (t0, tbound), y2, t_eval=t)
y0 = linear_ode_solution_0.y
y1 = linear_ode_solution_1.y
```

```python
plt.figure()
plt.plot(t, y0[0, :], 'r')
plt.plot(t, y1[0, :], 'k')
plt.ylim(0,5)
plt.title('RK45')
plt.xlabel('Time [s]')
plt.ylabel('Solution x')
plt.legend(['I.C. (1,1,1),  x(t)=exp(t)', 'I.C. (1,-1,1), x(t)=exp(-t)'])
plt.grid(True)
plt.show()
```
