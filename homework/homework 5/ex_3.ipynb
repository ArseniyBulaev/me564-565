**(b)**

```python
def f(x, y):
    return 20 - x**4 - y**2
```

```python
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(x, y)
zs = np.array(f(np.ravel(X), np.ravel(Y)))
Z = zs.reshape(X.shape)
```

```python
fig, ax = plt.subplots(subplot_kw={'projection':'3d'})
ax.plot_surface(X, Y, Z)
ax.plot([1, -1/np.sqrt(2)], [2,-1/np.sqrt(2)], [15, 0], 'r', markersize=1.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

```python
def gradT(x, y, z):
    return np.array([
        -np.sin(x) * np.cos(y) * np.sin(z),
        -np.cos(x) * np.sin(y) * np.sin(z),
        np.cos(x) * np.cos(y) * np.cos(z),
    ])
```

```python
gradT(2, 1, 1)
```
