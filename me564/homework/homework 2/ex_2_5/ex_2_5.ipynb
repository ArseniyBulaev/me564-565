```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
omega_1 = 1
omega_2 = 1.5
```

```python
def ODE_matrix(eps, omega_1, omega_2):
    A = np.array([
        [0, 0, 1, 0],
        [0, 0, 0, 1],
        [-omega_1**2 - eps, eps, 0, 0],
        [eps, -omega_2**2 - eps, 0, 0]
    ])
    return A

'''
d/dt[θ] = A [θ]
λ = a + ib
freq = b / 2π
'''
```

```python
def calculate_freq(epss):
    freq1 = np.zeros(len(epss))
    freq2 = np.zeros(len(epss))
    for i in range(len(epss)):
        lambdas = np.linalg.eig(ODE_matrix(epss[i], omega_1, omega_2)).eigenvalues
        freq1[i] = np.abs(lambdas.imag[0])/(2*np.pi)
        freq2[i] = np.abs(lambdas.imag[2])/(2*np.pi)
    return freq1, freq2
```

```python
def plot_freq(epss, freq1, freq2):
    plt.figure()
    plt.plot(epss, freq1)
    plt.plot(epss, freq2)
    plt.legend(["freq1", "freq2"])
    plt.xlabel('\u03B5')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()  
```

```python
epss = np.arange(0,0.5, 0.005)
freq1, freq2 = calculate_freq(epss)
plot_freq(epss, freq1, freq2)
```

```python
epss = np.arange(-2,2, 0.001)
freq1, freq2 = calculate_freq(epss)
plot_freq(epss, freq1, freq2)
for i in range(len(epss)):
    print("eps={}".format(epss[i]), freq1[i], freq2[i])
```
