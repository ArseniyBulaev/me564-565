```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

Exercise 3.1

```python
t0 = 0
tbound = 100
first_step = 0.01
max_step = first_step
t = np.arange(t0, tbound, first_step)
boundary_conditions = [0.5, -1]
A = [
    [0, 1],
    [-6, -5]
]
```

(a)

```python
def analytic_sol_a(t):
    return 0.5 * np.exp(-2 * t)

def  dxdt_a(t,X):
              return A @ X
```

```python
ode_solution_a = sp.integrate.solve_ivp(dxdt_a, (t0, tbound), boundary_conditions, t_eval=t)
x_a = ode_solution_a.y
```

```python
plt.figure()
plt.plot(t, x_a[0, :], 'r')
plt.plot(t, analytic_sol_a(t), 'g')
plt.xlabel('Time [s]')
plt.ylabel('Solution x')
plt.legend(['numeric', 'analytic'])
plt.grid(True)
plt.show()
```

(b)

```python
def analytic_sol_b(t):
    return -0.5 * np.exp(-2 * t) + 0.5 * np.exp(-3*t) + 0.5 * np.exp(-t)

def  dxdt_b(t,X):
              return A @ X + np.array([0, np.exp(-t)]).T
```

```python
ode_solution_b = sp.integrate.solve_ivp(dxdt_b, (t0, tbound), boundary_conditions, t_eval=t)
x_b = ode_solution_b.y
```

```python
plt.figure()
plt.plot(t, x_b[0, :], 'r')
plt.plot(t, analytic_sol_b(t), 'g')
plt.xlabel('Time [s]')
plt.ylabel('Solution x')
plt.legend(['numeric', 'analytic'])
plt.grid(True)
plt.show()
```

```python
def analytic_sol_c(t):
    return -39 / 2.0 * np.exp(-2 * t) + 15 * np.exp(-3*t) + 5*np.cos(t) + 5 * np.sin(t)

def  dxdt_c(t,X):
              return A @ X + np.array([0, 50*np.cos(t)]).T
```
